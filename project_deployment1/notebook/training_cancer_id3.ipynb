Library

```python
import pandas as pd
from sklearn import tree
from sklearn import metrics
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import joblib
```

Read data

```python
df_cancer = pd.read_csv('Cancer.csv')
df_cancer
```

```python
df_cancer['diagnosis'].value_counts()
```

```python
df_cancer['diagnosis'] = df_cancer['diagnosis'].map({'M': 1, 'B': 0})
df_cancer['diagnosis'].value_counts()
```

Independent Dependent Variable

```python
x = df_cancer.iloc[:,2:32].values
y = df_cancer.iloc[:,1].values
```

Split Train Test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

Create Model & Training 

```python
tree_clf = tree.DecisionTreeClassifier()
tree_clf = tree_clf.fit(x_train,y_train)
tree_clf
```

Testing Model 

```python
y_prediksi = tree_clf.predict(x_test)
y_prediksi
```

Evaluasi Model

```python
print("Test accuration : ", metrics.accuracy_score(y_test,y_prediksi))
ConfusionMatrixDisplay.from_predictions(y_test,y_prediksi)
```

Visualisasi Decision Tree

```python
fn = ['radius_mean','texture_mean','perimeter_mean','area_mean','smoothness_mean','compactness_mean','concavity_mean','concave points_mean','symmetry_mean','fractal_dimension_mean','radius_se','texture_se','perimeter_se','area_se','smoothness_se','compactness_se','concavity_se','concave points_se','symmetry_se','fractal_dimension_se','radius_worst','texture_worst','perimeter_worst','area_worst','smoothness_worst','compactness_worst','concavity_worst','concave points_worst','symmetry_worst','fractal_dimension_worst']
cn = ['B','M']
fig = plt.figure(figsize=(20,15))
pict_tree = tree.plot_tree(tree_clf,feature_names=fn,class_names=cn,filled=True)
```

```python
joblib.dump(tree_clf, '../models/cancer_id3.joblib')
```
