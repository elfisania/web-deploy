Library

```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
import joblib 
```

Read Data

```python
df_cancer = pd.read_csv('Cancer.csv')
df_cancer
```

```python
df_cancer.info()
```

```python
df_cancer['diagnosis'].value_counts()
```

```python
df_cancer['diagnosis'] = df_cancer['diagnosis'].map({'M': 1, 'B': 0})
df_cancer['diagnosis'].value_counts()
```

Independent Dependent Variable

```python
x = df_cancer.iloc[:,2:32].values
y = df_cancer.iloc[:,1].values
```

Split Train Test

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

Create Model & Training

```python
nb_clf = GaussianNB()
nb_clf.fit(x_train,y_train)
```

Testing Model

```python
y_prediksi = nb_clf.predict(x_test)
y_prediksi
```

Evaluasi Model

```python
print("Test accuration : ", metrics.accuracy_score(y_test,y_prediksi))
print(metrics.classification_report(y_test,y_prediksi))
ConfusionMatrixDisplay.from_predictions(y_test,y_prediksi)
```

Save Model

```python
joblib.dump(nb_clf, '../models/cancer_nb.joblib')
```
